```python
print("2.d") #Manually create a dictionary that has a single level and contains each patient as the key and the year as the value. Set Mochi's year to 2019.'

patients_dict = {"Kinoko": 2021,
                "Dango": 2019,
                "Mochi": 2019}
print(patients_dict)
```
