```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Getting the data
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'
covid_df = pd.read_csv(url, index_col=0)
covid_df.head()
```

```python
#Vis1 - Create a visualization that shows all of the counties in Utah as a time series,
#similar to the one shown in slide 22 during the lecture. The graphic should
#-Show cases over time
#-Have all counties plotted in a background color (something like grey)
#-Have a single county plotted in a contrasting color (something not grey)
#-Have well formatted dates as the X axis
```

```python
#Getting Utah data
Utah_data = covid_df[covid_df['Province_State'] == 'Utah']
Utah_data.head(10)
```

```python
import matplotlib.dates as mdates  # module for formatting dates
```

```python
# Convert columns to datetime
xpoints = pd.to_datetime(Utah_data.columns[11:], format='%m/%d/%y') #I ran it earlier and I had to indicate the date format to work faster

# Set up the plot
plt.figure(figsize=(12, 6))

# Format the axis to labels properly
font1 = {'family':'serif','color':'black','size':12}
font2 = {'family':'serif','color':'darkred','size':14}
font3 = {'family':'serif','color':'black','size':10}

#Formatting labels, fonts and title
plt.xlabel('Date', fontdict=font1)
plt.ylabel('Confirmed Cases', fontdict=font1)
plt.title('COVID-19 Confirmed Cases in Utah Counties', fontdict=font2)
plt.xticks(rotation=45, fontsize=10, fontfamily='serif')

# Set the data to be plotted per row (or per county)
for index, row in Utah_data.iterrows():
    ypoints = row[11:]  # Get the time series data for the county
    if row['Admin2'] == 'Salt Lake':
        plt.plot(xpoints, ypoints, color='red', label='Salt Lake County', linewidth=2.5)
    else:
        plt.plot(xpoints, ypoints, color='grey', alpha=0.5) 

# Formatting date on x-axis
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m')) #so it gets printed as year-month
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3)) #so it shows every 3 months

# Add legend
plt.legend()
# Show the plot
plt.show()

plt.tight_layout()
```

```python
#Vis2 - Create a visualization that shows the contrast between the county in Utah with
#the most cases to date to a county in Florida with the most cases to date.
#The graphic should:
#-Have only two counties plotted
#-Highlight the difference between the two comparison counties
#You may use any style of graphic you like as long as it is effective (dense)
#and readable
```

```python
#Get the Florida data
Florida_data = covid_df[covid_df['Province_State'] == 'Florida']
Florida_data.head(10)
```

```python
#Discover the county with the most cases in Utah
Utah_data['latest_cases'] = Utah_data[Utah_data.columns[-1]] # Get the latest cases for each county
max_utah_county = Utah_data.loc[Utah_data['latest_cases'].idxmax()]
print("Utah County with most cases:", max_utah_county['Admin2'], "Cases:", max_utah_county['latest_cases'])
```

```python
#Discover the county with the most cases in Florida
Florida_data['latest_cases'] = Florida_data[Florida_data.columns[-1]] # Get the latest cases for each county
max_florida_county = Florida_data.loc[Florida_data['latest_cases'].idxmax()]
print("Florida County with most cases:", max_florida_county['Admin2'], "Cases:", max_florida_county['latest_cases'])
```

```python
most_cases_utah_xpoint = max_utah_county['latest_cases']
most_cases_florida_xpoint = max_florida_county['latest_cases']
counties = [max_utah_county['Admin2'], max_florida_county['Admin2']]
cases = [most_cases_utah_xpoint, most_cases_florida_xpoint]
colors = ['blue', 'orange']
plt.figure(figsize=(8, 6))
bars = plt.bar(counties, cases, color=colors)
plt.xlabel('County', fontdict=font1)
plt.ylabel('Confirmed Cases', fontdict=font1)
plt.title('Comparison of COVID-19 Confirmed Cases: Utah vs Florida', fontdict=font2)

plt.show()
plt.tight_layout()
```

```python
#Vis 3 - #Create a visualization that shows BOTH the running total of cases for a single
#county AND the daily new cases. The graphic should:
#-Use two y-axes (https://matplotlib.org/stable/gallery/subplots_axes_and_figures/two_scales.html)
#-Use color to contrast the two series being plotted
#-Have well formatted dates as the X axis
```

```python
#Using the data from utah, but not the last column to not mess up the calculations
if 'latest_cases' in Utah_data.columns:
    Utah_data = Utah_data.drop(columns=['latest_cases'])

#Prepare the Data to grab only SLC
county_name = 'Salt Lake'

#Get the row for the specific county
county_row = Utah_data[Utah_data['Admin2'] == county_name].iloc[0]

#Extract the time series data (cols 11 onwards) from said row
cumulative_cases = county_row[11:]

#Create the 'Daily New Cases'. Will calculate the differences
daily_new_cases = cumulative_cases.diff().fillna(0)

#Get the Dates
dates = pd.to_datetime(Utah_data.columns[11:], format='%m/%d/%y')

#Create the plot
fig, ax1 = plt.subplots(figsize=(12, 6))

#Plot 1: Cumulative Cases (Left Axis)
color_total = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Total Cumulative Cases', color=color_total, fontsize=12)
ax1.plot(dates, cumulative_cases, color=color_total, linewidth=3)
ax1.tick_params(axis='y', labelcolor=color_total)

#Plot 2: Daily New Cases (Right Axis)
ax2 = ax1.twinx()
color_daily = 'tab:blue'
ax2.set_ylabel('Daily New Cases', color=color_daily, fontsize=12)
ax2.bar(dates, daily_new_cases, color=color_daily, alpha=0.3, width=1.0)
ax2.tick_params(axis='y', labelcolor=color_daily)

#Formatting
plt.title(f'COVID-19 Dynamics in {county_name} County: Total vs Daily Cases', fontsize=14)

ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

fig.tight_layout()
plt.show()
```

```python
#Vis 4 - Create a visualization that shows a stacked bar chart of county contributions
#to a given state's total cases. You may choose any state (or states).
#(https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_stacked.#html#sphx-glr-gallery-lines-bars-and-markers-bar-stacked-py)
#The graphic should:
#-Have a single column delineate a state
#-Have each 'slice' or column compontent represent a county
```

```python
#Selecting the states
states_of_interest = ['Hawaii', 'Delaware']
subset_vis4 = covid_df[covid_df['Province_State'].isin(states_of_interest)].copy()
latest_date_vis4 = subset_vis4.columns[-1]

#Taking just the data from the counties we selected
vis4_plot = subset_vis4.pivot_table(index = 'Province_State', columns = 'Admin2', values = latest_date_vis4, aggfunc = 'sum')

#creating the graph
ax = vis4_plot.plot(kind='bar', stacked=True, figsize=(10, 8), colormap='tab20')

#adding titles for understanding
plt.title('COVID-19 Total Cases: Hawaii vs. Delaware', fontsize=16)
plt.ylabel('Total Confirmed Cases')
plt.xlabel('State')

#Move the legend outside, mostly upper left
plt.legend(title='Counties', bbox_to_anchor=(1.05, 1), loc='upper left')

#finally plotting
plt.xticks(rotation=0) # Keep state names horizontal
plt.tight_layout()
plt.show()
```
