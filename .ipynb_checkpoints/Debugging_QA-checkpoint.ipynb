```python
def wrong_add_function(arg1,arg2):
   '''
   The function takes in two lists of integers, then it adds
   all of arg2 to each item of arg1.
   
   Example:
      > wrong_add_function([1,2,3],[1,1,1])
      > [6,9,12]

   whereas the expected correct answer is, [2,3,4]

   Parameters
   ----------
   arg1 : list
      list of integers.
   arg2 : list
      list of integers.

   Returns
   -------
   arg1 : list
      Elements of arg1, with each element having had the contents of 
      arg2 added to it.

   '''
   arg1_index=0
   while arg1_index < len(arg1):
      arg_2_sum = 0
      for arg2_elements in arg2:
         arg_2_sum = sum([arg1[arg1_index]+i for i in arg2])
      arg1[arg1_index]=arg_2_sum  
      arg1_index+=1
   return arg1
```

```python
def correct_add_function(arg1,arg2):
   '''
   Updated function to take in two lists of integers, then to add
   all of arg2 to each item of arg1.
   
   Example:
      > correct_add_function([1,2,3],[1,1,1])
      > [2,3,4]

   Parameters
   ----------
   arg1 : list
      list of integers.
   arg2 : list
      list of integers.

   Returns
   -------
   arg1 : list
      Elements of arg1, with each element having had the contents of 
      arg2 added to it.

   '''
   arg1_index=0
   while arg1_index < len(arg1):
      arg_2_sum = 0
      for arg2_elements in arg2:
         arg_2_sum = arg1[arg1_index] + arg2[arg1_index] #changing here to get each index to add against each other.
      arg1[arg1_index]=arg_2_sum  
      arg1_index+=1
   return arg1
```

```python
def wrong_add_function(arg1,arg2):
   '''
   The function takes in two lists of integers, then it adds
   all of arg2 to each item of arg1.
   
   Example:
      > wrong_add_function([1,2,3],[1,1,1])
      > [4,5,6]
   
   If the lists are lists of strings, concatenate them
   Example:
      > wrong_add_function(['1','2','3'],['1','1','1'])
      > ['1111','2111','3111']
   Parameters
   ----------
   arg1 : list
      list of integers.
   arg2 : list
      list of integers.

   Returns
   -------
   arg1 : list
      Elements of arg1, with each element having had the contents of 
      arg2 added to it.

   '''
   #numeric section
   if sum([type(i)==int for i in arg1])==len(arg1) and \
      sum([type(i)==int for i in arg2])==len(arg2):
         arg1_index=0
         while arg1_index < len(arg1):
            arg_2_sum = 0
            for arg2_elements in arg2:
               arg_2_sum = arg1[arg1_index] + arg2[arg1_index] #correcting the code to make it run as question 1
            arg1[arg1_index]=arg_2_sum  
            arg1_index+=1
         return arg1
   #string section
   elif sum([type(i)==str for i in arg1])==len(arg1) and \
      sum([type(i)==str for i in arg2])==len(arg2):
         arg1_index=0
         while arg1_index < len(arg1):
            arg_2_sum = ''
            for arg2_elements in arg2:
               arg_2_sum += arg2_elements
            arg1[arg1_index]=arg1[arg1_index]+str(arg_2_sum)
            arg1_index+=1
         return arg1
```

```python
def exception_add_function(arg1, arg2):
    '''
    The function takes in two lists of integers, then it adds
    all of arg2 to each item of arg1.
    
    Example:
       > wrong_add_function([1,2,3],[1,1,1])
       > [4,5,6]
    
    If the lists are lists of strings, concatenate them
    Example:
       > wrong_add_function(['1','2','3'],['1','1','1'])
       > ['1111','2111','3111']
    Parameters
    ----------
    arg1 : list
       list of integers.
    arg2 : list
       list of integers.

    Returns
    -------
    arg1 : list
       Elements of arg1, with each element having had the contents of 
       arg2 added to it.

    '''

    try:
        #now, if the try did not "work", then it runs the except block.
        #numeric section
        if sum([type(i)==int for i in arg1])==len(arg1) and \
          sum([type(i)==int for i in arg2])==len(arg2):
             arg1_index=0
             while arg1_index < len(arg1):
                arg_2_sum = 0
                for arg2_elements in arg2:
                   arg_2_sum = sum([arg1[arg1_index]+i for i in arg2])
                arg1[arg1_index]=arg_2_sum  
                arg1_index+=1
             return arg1
       #string section
        elif sum([type(i)==str for i in arg1])==len(arg1) and \
          sum([type(i)==str for i in arg2])==len(arg2):
             arg1_index=0
             while arg1_index < len(arg1):
                arg_2_sum = ''
                for arg2_elements in arg2:
                   arg_2_sum += arg2_elements
                arg1[arg1_index]=arg1[arg1_index]+str(arg_2_sum)
                arg1_index+=1
             return arg1
        else:# If neither of the above conditions were met, we will raise an error to force an except.
            raise TypeError("Input lists contain mixed or unsupported types.")

    except TypeError: #Adding the try function to create a check for types before running
        input_type = type(arg1[0]) #If the first element of the arg1 is a string, the rest should be. Same if it is an integer.
        # Check arg1 for unmatched inputs.
        for index, element in enumerate(arg1): #easiest way to get the elements and their respective indexes
            if not isinstance(element, input_type): #learned in advanced logic homework. Checking for the types.
                return (f"Your input argument 1 at element[{index}]is not of the expected type. Please change this and rerun.") #If the type does not\
                #match, then produce the message asked in this homework

        # Check arg2 in the same way as arg1.
        for index, element in enumerate(arg2):
            if not isinstance(element, input_type): #same as above
                return (f"Your input argument 2 at element[{index}]is not of the expected type. Please change this and rerun.")
```

```python
def correction_add_function(arg1,arg2):
    try:
        #Trying the wrong_add_function that I've previously corrected
        result = wrong_add_function(arg1, arg2) #saving as a result since I will not be able to have a result to return if this code works
        
        # If block to check if the result happened or not
        if result is not None: #if the function worked, the result is printed
            return result
        # Otherwise, if the result was None, raise the error.
        else:
            raise TypeError
        
    except TypeError: #Catch the error raised above and preparing to correct everything so the code runs
        correct_arg1 = [str(element) for element in arg1] #taking all of the elements so it can run through the string block
        correct_arg2 = [str(element) for element in arg2]
        return wrong_add_function(correct_arg1, correct_arg2) #run again, but the corrected inputs

    finally:
        print("we made sure the function will work ;)")
```
