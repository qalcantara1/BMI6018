```python
print("3.1 - write a function, that takes in the array as an argument, and outputs 10 values from the dataset, spaced out by indexes that are 25 apart (ie 0, 25, 50, etc)")
```

```python
def spaced_25_function(input_array):
    i = 0 #set up a start point
    for index in range(len(input_array)):
        if index % 25 == 0: #spaced by 25
            print(input_array[index])
            i += 1 #so we can add to the count
            if i >= 10: #make sure only 10 items are printed
                break
```

```python
spaced_25_function(influenza_genome)
```

```python
print("3.2 - write a function that takes in the dataset as an argument and outputs 20 values from the dataset, spaced out by indexes that are y apart (ie you can decide how far apart they should be iterated as long as they dont exceed the length of the dataset)")
```

```python
def print_spaced_values(input_array, spacing): #I will not decide the spacing, I will let the user define it
    result = [] #adding to a list, since we are using it later
    for i in range(20): # make sure there are 20 output values
        index = i * spacing
        if index < len(input_array): #so it doesn't go out of bounds
            result.append(input_array[index])
        else:
            break

    return result
```

```python
print_spaced_values(influenza_genome, 20)
```

```python
print("3.3 - write a function that takes the output from the function from 3.2 as an argument, then only prints out every other item (ie there should only be 10 outputs)")
```

```python
def every_other_value_function(data):
    previous_result = print_spaced_values(data, 2) #take up the input from 3.2 function, I set it up for the spacing to be 2
    for item in previous_result[::2]: #since we have only 20 elements and we are reducing to 10, I can just divide by 2.
        print(item)
```

```python
every_other_value_function(influenza_genome)
```

```python
print("4.1 - Write a function that implements the code from problem 1.4, then implements the code from problem 2.3.")
```

```python
def narrowing_inputs(data):
    between_list = [] #creating empty list for the processing data
    final_data = [] #placeholder for the final data
    for index, item in enumerate(data[200:301]): #for all elements, including 300th
            if item % 3 == 0 and item % 5 == 0: #divisible by both, as required in 1.4
                between_list.append(10)
            else:
                between_list.append(item)
                
    for index in range(34, 38): #get the indexes that would be equivalent to 234 and 238 (here we already start at 200)
        item = between_list[index]
        if index == 36: #take only the 236th item (or here it would be the 36th)
            # Loop 7 times to print the item
            for i in range(7):
                final_data.append(item)
        else:
            final_data.append(item)
    return final_data
```

```python
narrowing_inputs(influenza_genome)
```

```python
def modify_function(data):
    modified_genome = data
    for index in range(200, 301):
        if index < len(modified_genome): #so it doesn't go over bounds
            item = modified_genome[index]
            if item % 3 == 0 and item % 5 == 0:
                modified_genome[index] = 10 #modifying the data
    for index in range(234, 238):
        if index < len(modified_genome):
            item = modified_genome[index]             
            if index == 236: # If the index is 236, print the item 7 times
                for i in range(7):
                    print(item) #the question asked to print it
            else:
                print(item)
```
