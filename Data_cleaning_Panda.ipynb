```python
#Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importing data
flights_data = pd.read_csv('flights.csv') #I kept the data in the same folder to help with the process
weather_data_pd = pd.read_csv('weather.csv')
weather_data_np = weather_data_pd.to_numpy()
```

```python
#Checking the data
flights_data.head(10)
```

```python
#Question 1 - How many flights were there from JFK to SLC? Int
q_1 = len(flights_data[(flights_data['origin'] == 'JFK') & (flights_data['dest'] == 'SLC')]) #use len and a filtering method
print(q_1)
```

```python
#Question 2 - How many airlines fly to SLC? Should be int
q_2 = flights_data[flights_data['dest'] == 'SLC']['carrier'].nunique()
print(q_2)
```

```python
#Question 3 - What is the average arrival delay for flights to RDU? float
q_3= flights_data[flights_data['dest'] == 'RDU']['arr_delay'].mean()
print(q_3)
```

```python
#Question 4 - What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float
total_sea_flights = len(flights_data[flights_data['dest'] == 'SEA']) #get the total
nyc_to_sea_flights = len(flights_data[
    (flights_data['dest'] == 'SEA') & 
    (flights_data['origin'].isin(['LGA', 'JFK'])) #specific for LGA and JFK
])

q_4 = nyc_to_sea_flights / total_sea_flights

q_4 #Print the float proportion
```

```python
#Question 5 - Which date has the largest average depature delay? Pd slice with date and float
#please make date a column. Preferred format is 2013/1/1 (y/m/d)
avg_dep_delay_by_date = flights_data.groupby(['year', 'month', 'day'])['dep_delay'].mean()
# .nlargest(1) returns a Series (a "pd slice") with the index (date) and the value.
q_5 = avg_dep_delay_by_date.nlargest(1) #Pd slice from what I googled it

#result
q_5
```

```python
#Question 6 - Which date has the largest average arrival delay? pd slice with date and float
#same as aboe, but different parameter
avg_arrival_delay_by_date = flights_data.groupby(['year', 'month', 'day'])['arr_delay'].mean()
q_6 = avg_arrival_delay_by_date.nlargest(1)
q_6 #print
```

```python
#Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed
#speed = distance/airtime
nyc_flights = flights_data[
    (flights_data['origin'].isin(['LGA', 'JFK'])) &
    (flights_data['year'] == 2013)
].copy() #learned I have to use copy to not ruin the data. Getting only thos airports and that date
```

```python
#distance by airtime per hour
nyc_flights['speed'] = nyc_flights['distance'] / (nyc_flights['air_time'] / 60)  

# Find the row with the maximum speed
fastest_flight = nyc_flights.loc[nyc_flights['speed'].idxmax()] #getting the maximum

# Select only the tailnumber and speed as requested
q_7 = fastest_flight[['tailnum', 'speed']]

q_7
```
