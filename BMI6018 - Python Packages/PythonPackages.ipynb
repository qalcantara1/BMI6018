```python
import numpy as np
import pandas as pd 
from scipy import stats
```

```python
#Getting the data
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'
covid_df = pd.read_csv(url, index_col=0)
covid_df.head()
```

```python
#Getting Utah data
Utah_data = covid_df[covid_df['Province_State'] == 'Utah']
Utah_data.head(10)
```

```python
#Get the Florida data
Florida_data = covid_df[covid_df['Province_State'] == 'Florida']
Florida_data.head(10)
```

```python
#Discover the county with the most cases in Utah
Utah_data['latest_cases'] = Utah_data[Utah_data.columns[-1]] # Get the latest cases for each county
```

```python
#Getting the top 3 counties with the max cases in Utah using latest_cases column
top_3_utah = Utah_data.nlargest(3, 'latest_cases')
print("Top 3 Utah counties by latest cases:")
print(top_3_utah[['Admin2', 'latest_cases']])
```

```python
#Discover the county with the most cases in Florida
Florida_data['latest_cases'] = Florida_data[Florida_data.columns[-1]] # Get the latest cases for each county
```

```python
#Getting the top 3 counties with the max cases in Florida using latest_cases column
top_3_florida = Florida_data.nlargest(3, 'latest_cases')
print("Top 3 Florida counties by latest cases:")
print(top_3_florida[['Admin2', 'latest_cases']])
```

```python
#Using Scipy to perform a t-test comparing the latest case counts of the top 3 counties from each state
utah_top3_cases = top_3_utah['latest_cases'].values
florida_top3_cases = top_3_florida['latest_cases'].values

t_stat, p_value = stats.ttest_ind(utah_top3_cases, florida_top3_cases)
print(f"T-test results comparing top 3 counties (Utah vs Florida):")
print(f"Utah top 3 cases: {utah_top3_cases}")
print(f"Florida top 3 cases: {florida_top3_cases}")
print(f"T-statistic: {t_stat}, p-value: {p_value}")
```

```python
import matplotlib.pyplot as plt

# Create a bar plot comparing the top 3 counties from each state
fig, ax = plt.subplots(figsize=(10, 6))

# Get county names and cases
utah_counties = top_3_utah['Admin2'].values
utah_cases_vals = top_3_utah['latest_cases'].values
florida_counties = top_3_florida['Admin2'].values
florida_cases_vals = top_3_florida['latest_cases'].values

# X-axis positions
x_utah = range(len(utah_counties))
x_florida = range(len(utah_counties), len(utah_counties) + len(florida_counties))

# Plot bars
ax.bar(x_utah, utah_cases_vals, label='Utah', alpha=0.7, color='blue')
ax.bar(x_florida, florida_cases_vals, label='Florida', alpha=0.7, color='red')

# Set x-axis labels
all_counties = list(utah_counties) + list(florida_counties)
ax.set_xticks(list(x_utah) + list(x_florida))
ax.set_xticklabels(all_counties, rotation=45, ha='right')

# Add labels and title
ax.set_ylabel('Latest Cases', fontsize=12)
ax.set_xlabel('County', fontsize=12)
ax.set_title('Top 3 Counties by Cases: Utah vs Florida', fontsize=14, fontweight='bold')

# Add p-value text on the plot
ax.text(0.5, 0.95, f'p-value = {p_value:.4f}', 
        transform=ax.transAxes, fontsize=12, fontweight='bold',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8),
        verticalalignment='top', horizontalalignment='center')

ax.legend()
plt.tight_layout()
plt.show()
```
